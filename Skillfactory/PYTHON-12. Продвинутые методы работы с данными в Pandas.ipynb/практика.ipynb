```python
import pandas as pd
import re
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

```python
ratings_movies = pd.read_csv('data/ratings_movies.csv')
#ratings_movies.info()
ratings_movies['year_release'] = ratings_movies['title'].apply(get_year_release)
#mask1 = ratings_movies['year_release'] == None
#display(ratings_movies[mask1].loc[:, ['year_release', 'title', 'rating']])

ratings_movies.info()
#display(ratings_movies['year_release'].sort_values)
```

```python
# Информация в таблицах представлена в виде следующих столбцов:

#item_id — идентификатор модели;
#vendor — производитель модели;
#stock_count — имеющееся на складе количество данных моделей (в штуках);
#purchase_id — идентификатор покупки;
#price — стоимость модели в покупке.

# 1 Сформируйте DataFrame merged,
merged = items_df.join(
    purchase_df.set_index('item_id'),
    on='item_id',
    how='inner'
)
# в котором в результате объединения purchase_df и items_df останутся модели, 
# которые учтены на складе и имели продажи.
#
display(merged) 
# 2 Из таблицы merged найдите суммарную выручку,
# которую можно было бы получить от продажи всех товаров,
# которые учтены на складе и имели продажи. 
# 
# Результат занесите в переменную income.


series = merged['stock_count']*merged['price']
income = series.sum() 
```
